# Предобработка данных

### Шкалирование всех признаков

Шкалирование признаков нужно, для того, чтобы все признаки были в одном маштабе. Другими словами мы обезразмериваем все признаки.

Зачем это нужно? На примере линейной регресии, мы сумируем все признаки с некоторыми весами $\textbf{w}$. Все признаки имеют разные размерности, а еще хуже они все разные по порядку величины. Тоесть все $\textbf{w}$ будут тоже разными по размерностям и порядкам величины. Для этого все признаки нужно отнормировать или отшкалировать.

Шкалирование происходит следующим образом. Сначала от каждого признака вычетается его среднее значение, чтобы после нормировки все признаки имели среднее 0. После этого каждый признак делиться на дисперсию соответсвующего признака (под дисперсией можно понимать насколько все признаки в среднем отклоняются от среднего значения).

Замечание: шкалировать нужно только признаки, ответы шкалируються только в случае регресионной задачи.

Для шкалирование есть хорошая функция $\textbf{scale}$, которая делает шкалирование, но она работает только для $\textbf{numpy}$.


##### Сначала разобьем наш dataframe на признаки и ответы.

```python
fd_without_target = fd.drop(columns=['Type'])
```

```python
fd_without_target.head()
```

```python
fd_target = fd.drop(columns=fd.columns[1:])
```

```python
fd_target.head()
```

##### Теперь сделаем шкалирование

```python
fd_without_target_scale = fd_without_target - fd_without_target.mean()
```

```python
fd_without_target_scale.describe()
```

```python
fd_without_target_scale = fd_without_target_scale/fd_without_target.std()
```

```python
fd_without_target_scale.describe()
```

Как видно после шкалирование, получили, что все признаки имеют нулевое среднее и единичную дисперсию.

```python
fd_without_target_scale.boxplot(figsize=(25,10))
```

В pandas dataframe мы просто сделали это для себя, чтобы увидеть эти данные, теперь сделаем тоже самое для numpy массива, тоесть для нашего рабочего масива.

```python
fd_np.shape
```

```python
X = fd_np[:,1:]
y = np.reshape(fd_np[:,:1], -1)
```

Получили два массива, $\textbf{X}$ - массив признаков, а $\textbf{y}$ - массив ответов.

```python
print(X.shape, y.shape)
```

Теперь используем встроеную функцию которая шкалирует признаки для $numpy$ массива

```python
from sklearn.preprocessing import scale
```

```python
X = scale(X)
```

Посмотрим на значение среднего и дисперсии после нормировки.

```python
print('Среднее:')
print(X.mean(axis= 0))
print('Дисперсия:')
print(X.std(axis= 0))
```

# Обучение модели

Давайте теперь используем модель kNN для решения этой задачи.

Будем испольовать готовую библиотеку $\textbf{sklearn}$

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kNN = KNeighborsClassifier(n_neighbors=1)
```

```python
kNN.fit(X, y)
```

```python
y_answ = kNN.predict(X)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('Accuracy = ' + str(accuracy_score(y_true=y, y_pred=y_answ)))
```

Мы получили точность прогноза 1.0, тоесть мы очень точно определяем ответы. Но в чем подвох? А подвох в том, что модель видела все данные в моменте обучения. И выставив один ближайший сосед, она всегда выбирает себя в качестве ближайшего. Поэтому теперь мы разделим выборку на обучающую и тестовую. Тестовая выборка это выборка на которой тестируется модель, ее модель в момент обучения не видит.

# Обучение модели с более честным представлением

Для разделения выборки на обучающую и на тестовую используем стандартную функцию $\textbf{train_test_split}$, которая делит общую выборку на обучающую и тестовую в произвольном порядке, тоесть рандомно в заданом соотношении.

```python
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(0)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
kNN = KNeighborsClassifier(n_neighbors=1)
```

```python
kNN.fit(X_train, y_train)
```

```python
y_answ = kNN.predict(X_test)
```

```python
print('Accuracy = ' + str(accuracy_score(y_true=y_test, y_pred=y_answ)))
```

Теперь на заданом разбиении выборки, построим график зависимости качества прогноза на тесте, при изменении числа ближайших соседей.

```python
N = X_train.shape[0]
List_of_K = np.linspace(1, N, num = N, dtype=np.int64)
```

```python
List_of_accuracy = []

for k in List_of_K:
    kNN = KNeighborsClassifier(n_neighbors=k)
    kNN.fit(X_train, y_train)
    List_of_accuracy.append(accuracy_score(y_true=y_test, y_pred=kNN.predict(X_test)))
```

```python
plt.plot(List_of_K, List_of_accuracy)

plt.grid()
```

Как видно из графика, лучше всего взять малое количество ближайших соседей, порядка 7-10, тогда точность на тестовой выборки будет 1.0, и это уже при условии, что модель не видела этой тестовой выборки.
