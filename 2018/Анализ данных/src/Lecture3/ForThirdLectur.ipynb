```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2+x+1

def df(x):
    return 2*x+1
```

```python
x = np.linspace(-1,10, 100)
y = f(x)
```

```python
x_0 = 6
y_0 = f(x_0)
print(x_0, y_0)
```

```python
alpha = df(x_0)
b = y_0 - alpha*x_0

dy = alpha*x + b
print(alpha, b)
```

```python
plt.plot(x, y, label = 'y = x^2+x+1')
plt.plot(x, dy, label = 'y = y\'(6)x-35')


plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend(loc = 'best')

plt.savefig('Lecture_3_derivation.pdf')
plt.show()
```
