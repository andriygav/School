```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
N = 10
a = -3.5
b = 3.5

x = np.linspace(a, b, N, dtype=np.float64)
y = x**2 + np.random.normal(loc = 0, scale=0.5, size = N)

_x = np.linspace(a, b, N**2, dtype=np.float64)
_y = _x**2
```

```python
plt.plot(x, y, 'bo', label = 'Our Data')
plt.plot(_x, _y, 'r-', label = 'y = x^2')

plt.legend(loc='best')
plt.grid()

plt.savefig('Lecture_2_regress.pdf')
plt.show()
```

```python
N = 20

x1 = np.random.normal(loc = -1.0, scale = 1.0, size = N)
y1 = np.random.normal(loc = -1.0, scale = 1.0, size = N)

x2 = np.random.normal(loc = 1.0, scale = 1.0, size = N)
y2 = np.random.normal(loc = 1.0, scale = 1.0, size = N)
```

```python
plt.plot(x1, y1, 'bo', label = 'class 0')
plt.plot(x2, y2, 'ro', label = 'class 1')

plt.legend(loc='best')
plt.grid()

plt.savefig('Lecture_2_class.pdf')
plt.show()
```
